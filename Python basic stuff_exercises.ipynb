```python
#Varianta 2 - separare numere in alta lista w/ WHILE loop

lst2 = ['Maria', 100, 'Harry', 30, 10240000000000, 4, 5, 5, 5, 4, 44, 30, 3, 5]
numericList2=list()
count = 0

while count<len(lst2):
    if str(lst2[count]).isnumeric():
        numericList2.append(lst2[count])
        lst2.pop(count)
    else:
        count=count+1

print("numerele sunt: {}".format(numericList2))
print("celalalte elemente sunt: {}".format(lst2))
```

# Zip() function


```python
#Ex 1
lst1 = [11, 22, 31, 55]
lst2 = [1, 51, 2, 88]
tmp = []
cmb = zip(lst1,lst2)
for i, j in cmb:
    if i>j:
        tmp.append(i-j)
        i=i-1
    elif i<j:
        tmp.append(i*j)
        j = j-i
```

```python
# Ex 2 - Cauta intr-o lista, valorile care se potrivesc in range-uri definite in alte liste

sysName = ["s1","s2","s3","s4","xxx"]        
in1 = [0.6, 1.2, 1.6, 23.5]
in2 = [2.4, 2.1,  3.6, 56]

nsx = [1.3, 2, 24, 51, 2, 1.2, 1.5, 0.5, 0.4]

resultList = []
sxp = []

for p in nsx:
    for i,j,k in zip(sysName, in1, in2):
        if p>=j and p<=k:
            resultList.append([i, p])
            found = True
        else:
            resultList.append(["xxx", p])
            found = False
            foundUnk= True
        if found is True:
            if i not in sxp:
                sxp.append(i)
if foundUnk is True:
    sxp.append("xxx")
nameList, pList = zip(*resultList)
print("Found systems: {}".format(sxp))

finalResult = []
for x in sxp:
    tmp = []
    for j,k in zip(nameList, pList):
        if j == x:
            tmp.append(k)
    finalResult.append([x, tmp])
    
for i,j in finalResult:
    print("Found vals {}: {}".format(i,j))
```
